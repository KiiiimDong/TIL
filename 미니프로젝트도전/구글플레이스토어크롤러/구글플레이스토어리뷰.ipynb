```python
#Step 0. 필요한 모듈과 라이브러리를 로딩합니다.
import sys # 시스템
import os  # 시스템

import pandas as pd  # 판다스 : 데이터분석 라이브러리
import numpy as np   # 넘파이 : 숫자, 행렬 데이터 라이브러리
from datetime import datetime
from bs4 import BeautifulSoup    # html 데이터를 전처리
from selenium import webdriver   # 웹 브라우저 자동화
from selenium.webdriver.common.keys import Keys
import chromedriver_autoinstaller

import time    # 서버와 통신할 때 중간중간 시간 지연. 보통은 1초
from tqdm import tqdm_notebook   # for문 돌릴 때 진행상황을 %게이지로 알려준다.

# 워닝 무시
import warnings
warnings.filterwarnings('ignore')
```

```python
# Step 1. 크롬 웹브라우저 실행
chrome_path = chromedriver_autoinstaller.install()
driver = webdriver.Chrome(chrome_path)
# 구글 앱스토어 앱 페이지
driver.get('https://play.google.com/store/apps/details?id=com.cyworld.minihompy')
time.sleep(2)  # 2초간 정지
```

```python
os.makedirs('result', exist_ok=True)

driver.execute_script('window.scrollTo(0, 1000);')
time.sleep(3)
  
try:
    load_more = driver.find_element_by_css_selector('#yDmH0d > c-wiz.SSPGKf.Czez9d > div > div > div.tU8Y5c > div.wkMJlb.YWi3ub > div > div.qZmL0 > c-wiz:nth-child(4) > section > header > div > div:nth-child(2) > button > i').click()
except:
    print('Cannot find load more button...')
time.sleep(3)
```

```python
# create empty dataframe to store data
df = pd.DataFrame(columns=['name', 'ratings', 'date', 'comment'])

# ActionChains 를 사용하기 위해서.
from selenium.webdriver import ActionChains

for i in range(1,501):
    try :
        # id가 something 인 element 를 찾음
        some_tag = driver.find_element_by_css_selector(f'#yDmH0d > div.VfPpkd-Sx9Kwc.cC1eCc.UDxLd.PzCPDd.HQdjr.VfPpkd-Sx9Kwc-OWXEXe-FNFY6c > div.VfPpkd-wzTsW > div > div > div > div > div.fysCi > div > div:nth-child(1) > div:nth-child({i+2}) > header > div.YNR7H > div.gSGphe > div')
        
        # somthing element 까지 스크롤
        action = ActionChains(driver)
        action.move_to_element(some_tag).perform()
        time.sleep(2)
        # reviewer
        name = driver.find_element_by_css_selector(f'#yDmH0d > div.VfPpkd-Sx9Kwc.cC1eCc.UDxLd.PzCPDd.HQdjr.VfPpkd-Sx9Kwc-OWXEXe-FNFY6c > div.VfPpkd-wzTsW > div > div > div > div > div.fysCi > div > div:nth-child(1) > div:nth-child({i}) > header > div.YNR7H > div.gSGphe > div').text
        #print(name)

        # rating
        ratings = driver.find_element_by_css_selector(f'#yDmH0d > div.VfPpkd-Sx9Kwc.cC1eCc.UDxLd.PzCPDd.HQdjr.VfPpkd-Sx9Kwc-OWXEXe-FNFY6c > div.VfPpkd-wzTsW > div > div > div > div > div.fysCi > div > div:nth-child(1) > div:nth-child({i}) > header > div.Jx4nYe > div').get_attribute('aria-label').replace('별표 5개 만점에', '').replace('개를 받았습니다.', '').strip()
        #print(ratings)
    
        # review date
        date = driver.find_element_by_css_selector(f'#yDmH0d > div.VfPpkd-Sx9Kwc.cC1eCc.UDxLd.PzCPDd.HQdjr.VfPpkd-Sx9Kwc-OWXEXe-FNFY6c > div.VfPpkd-wzTsW > div > div > div > div > div.fysCi > div > div:nth-child(1) > div:nth-child({i}) > header > div.Jx4nYe > span').text
        #print(date)

        # helpful
        #helpful = driver.find_element_by_css_selector(f'#yDmH0d > div.VfPpkd-Sx9Kwc.cC1eCc.UDxLd.PzCPDd.HQdjr.VfPpkd-Sx9Kwc-OWXEXe-FNFY6c > div.VfPpkd-wzTsW > div > div > div > div > div.fysCi > div > div:nth-child(1) > div:nth-child({i}) > div:nth-child(3) > div').text.replace('사용자', '').replace('명이 이 리뷰가 유용하다고 평가함', '').strip()
        #print(helpful)

        # review text
        comment = driver.find_element_by_css_selector(f'#yDmH0d > div.VfPpkd-Sx9Kwc.cC1eCc.UDxLd.PzCPDd.HQdjr.VfPpkd-Sx9Kwc-OWXEXe-FNFY6c > div.VfPpkd-wzTsW > div > div > div > div > div.fysCi > div > div:nth-child(1) > div:nth-child({i}) > div.h3YV2d').text
        #print(comment)
        # append to dataframe
        df = df.append({'name': name,'ratings': ratings,'date': date,'comment': comment}, ignore_index=True)
        print(f'{i}번째리뷰  크롤링중')      
    except:
        print('error')

# finally save the dataframe into csv file
filename = datetime.now().strftime('result/%Y-%m-%d_%H-%M-%S.csv')
df.to_csv(filename, encoding='utf-8-sig', index=False)
driver.stop_client()
driver.close()

print('Done!')
```
