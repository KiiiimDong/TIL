```python
import matplotlib.pyplot as plt
x=[1,2,3]
y=[-4,-8,-6]
plt.plot(x,y)
plt.title("꺾은선 그래프",fontdict={'family':'Malgun Gothic','size':25})
#이거에만 대해서 적용 #한글로 전체 설정하면 마이너스 깨진다.
plt.xlabel('가로축',color="#00aa00",fontdict={'family':'Malgun Gothic','size':10},loc='left')
#가로 left,center,right
plt.ylabel('세로축',color="#0000aa",fontdict={'family':'Malgun Gothic','size':10},loc='bottom')
#세로 top,center,bottom
plt.show()
```

```python
x=[1,2,3]
y=[4,8,6]
plt.plot(x,y)
plt.title("꺾은선 그래프",fontdict={'family':'Malgun Gothic','size':25})
#이거에만 대해서 적용 #한글로 전체 설정하면 마이너스 깨진다.
plt.xlabel('가로축',color="#00aa00",fontdict={'family':'Malgun Gothic','size':10},loc='left')
#가로 left,center,right
plt.ylabel('세로축',color="#0000aa",fontdict={'family':'Malgun Gothic','size':10},loc='bottom')
#세로 top,center,bottom
plt.xticks=([1,2,3])
plt.yticks=([4,6,8])
plt.show()
```

```python
#그림생성
fig=plt.figure()
fig.set_size_inches(10,10)
import numpy as np
X_1=range(100)
Y_1=[np.cos(i) for i in X_1]

X_2=range(100)
Y_2=[np.sin(i) for i in X_2]
#그래프 생성
ax_1 = fig.add_subplot(1,2,1) #행, 열 , 위치
ax_2 = fig.add_subplot(1,2,2)
ax_1.plot(X_1,Y_1)
ax_2.plot(X_2,Y_2)
plt.show()
```

```python
x=np.linspace(-1,1,100)
y1=np.sin(x)
y2=np.cos(x)
y3=np.tan(x)
y4=np.exp(x)
plt.plot(x,y1,c='r',ls='dashed',label='sin')
plt.plot(x,y2,c='b',ls='dotted',label='cos')
plt.plot(x,y3,c='g',ls='solid',label='tan')
plt.plot(x,y4,c='c',label='exp',linewidth=5)
plt.legend(loc="best",shadow=True)
plt.show()
```

```python
x=np.linspace(-1,1,100)
y1=np.sin(x)
y2=np.cos(x)
y3=np.tan(x)
y4=np.exp(x)
plt.plot(x,y1,c='r',ls='dashed',label='sin',marker='o')
plt.plot(x,y2,c='b',ls='dotted',label='cos')
plt.plot(x,y3,c='g',ls='solid',label='tan')
plt.plot(x,y4,c='c',ls='None',label='exp',linewidth=5,marker='o')#ls=None도가능
plt.legend(loc="best",shadow=True)
plt.show()
```
