data 수집

```python
from konlpy.corpus import kolaw
t_data = kolaw.open('constitution.txt').read()
type(t_data)
```

전처리 - (형태소 분석기 이용)

```python
from konlpy.tag import Okt
tw = Okt()
n_t_data = tw.nouns(t_data)
type(n_t_data)
```

시각화 작업을 위한 NLTK Text 클래스 이용 정리

```python
from nltk import Text
ck_data = Text(n_t_data,name='kolaw')
ck_data
```

정리된 내용을 이용하여 data 분석

```python
from wordcloud import WordCloud
f = r'C:\Windows\Fonts\NanumGothic.ttf' # r
wc = WordCloud(font_path=f, width=1000, height=600, background_color='white')
plt.imshow(wc.generate_from_frequencies(ck_data.vocab()))
plt.axis('off')
```
