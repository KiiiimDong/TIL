```python
from sklearn.datasets import *
import pandas as pd
cancer = load_breast_cancer()

#cancer.target_names #['malignant', 'benign']
#import numpy as np
#np.unique(data['Y'])

data = pd.DataFrame(cancer.data, columns=cancer.feature_names)
sy = pd.Series(cancer.target, dtype="category")
sy = sy.cat.rename_categories(cancer.target_names)
data['class'] = sy
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
for i in range(6):
    k = cancer.feature_names[5*i:5+(5*i)]
    sns.pairplot(vars=k,data=data,hue='class')
    plt.show()
```

```python
sns.pairplot(vars=["worst radius", "worst texture", "worst perimeter", "worst area"], 
             hue="class", data=data)
```

```python
X_data = data[['worst radius','worst texture','worst perimeter','worst area']].values
```

```python
Y_data = data['class'].replace(['malignant', 'benign'],[0,1])
```

```python
from tensorflow.keras.utils import to_categorical
Y_data = to_categorical(Y_data)
Y_data
```

```python
from sklearn.model_selection import train_test_split
t_x,tt_x,t_y,tt_y = train_test_split(X_data,Y_data,random_state=3)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
m = Sequential()
m.add(Dense(2, input_dim=4, activation='sigmoid'))
m.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['accuracy'])
hy = m.fit(t_x,t_y,epochs=200,batch_size=1,validation_data=(tt_x,tt_y))
```

```python
import matplotlib.pyplot as plt
ec = range(1,len(hy.history['accuracy'])+1)
plt.plot(ec,hy.history['loss'])
plt.plot(ec,hy.history['val_loss'])
```
